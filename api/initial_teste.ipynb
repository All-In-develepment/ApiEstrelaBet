```python
import requests
import pymongo
from pymongo import MongoClient
from dotenv import load_dotenv, find_dotenv
import os
import datetime
```

```python
# Carregar as variáveis do arquivo .env
_ = load_dotenv(find_dotenv())

# Configurações do MongoDB
username = os.getenv("MONGO_ADMIN")
password = os.getenv("MONGO_PASSWORD")
mongo_client = MongoClient(f"mongodb://{username}:{password}@localhost:27017/")
db = mongo_client["easy_analytics_db"]
collection = db["kiron_partida"]
```

```python
# URL da API
url = "https://www.plataforma.easycoanalytics.com.br/api-easy-analytics/api.php?action=getKironPartida"
```

```python
# Fazer a requisição para a API
response = requests.get(url).json()
```

```python
for partida in response['retorno']:
    print(partida)
```

```python
query = {
    "league": "ItalianFastLeagueFootballSingleMatch",
    "hour_match": "14:35",
    "data_match": "03/09/2024"
}
```

```python
# Executar a consulta
result = collection.find_one(query)
```

```python
for partida in response['retorno'][0:3]:
    query = {
        "league": partida['league'],
        "hour_match": partida['hour_match'],
        "data_match": partida['data_match'],
    }
    result = collection.find_one(query)
    partida['timestamp'] = datetime.datetime(int(partida['data_match'].split('/')[2]), int(partida['data_match'].split('/')[1]), int(partida['data_match'].split('/')[0]), int(partida['hora']), int(partida['minuto']))
    last_game_spanish = partida
    if result == None:
        collection.insert_one(partida)
        print('Partida inserida com sucesso')
```
